To find the inverse of a matrix in Python, you can use the `numpy.linalg.inv()` function from the NumPy library. First, you need to define your matrix as a NumPy array.

```python
import numpy as np

# Define a sample square matrix
A = np.array([[1, 2],
              [3, 4]])

print("Original matrix A:")
print(A)

# Calculate the inverse of the matrix
# Note: A matrix must be square and non-singular (determinant is not zero) to have an inverse.
try:
    A_inv = np.linalg.inv(A)
    print("\nInverse of matrix A:")
    print(A_inv)

    # Verify the inverse (A * A_inv should be close to identity matrix)
    identity_check = np.dot(A, A_inv)
    print("\nVerification (A * A_inv):")
    print(identity_check)

except np.linalg.LinAlgError:
    print("\nThe matrix is singular and does not have an inverse.")
```
