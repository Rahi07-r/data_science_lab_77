```python
score = 75

if score >= 60:
    print("You passed!")
else:
    print("You failed.")

# You can also use elif for multiple conditions
grade = 'B'

if grade == 'A':
    print("Excellent!")
elif grade == 'B':
    print("Good job!")
elif grade == 'C':
    print("You can do better.")
else:
    print("Please study harder.")
```

This code demonstrates how to use `if`, `elif` (else if), and `else` statements to execute different blocks of code based on certain conditions. In the first example, it checks if a student's score is passing. In the second, it assigns a message based on a letter grade.
