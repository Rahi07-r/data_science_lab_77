```python
import numpy as np

# Define the coefficients matrix A
A = np.array([[2, 3], [4, 5]])

# Define the constants vector B
B = np.array([10, 20])

# Solve the system of equations
solution = np.linalg.solve(A, B)

# Print the solution
print(f"The solution is x = {solution[0]:.2f} and y = {solution[1]:.2f}")
```
