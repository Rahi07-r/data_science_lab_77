```python
# Define a 3x3 matrix
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

# Calculate the determinant
# Formula for a 3x3 matrix determinant:
# det(A) = a(ei - fh) - b(di - fg) + c(dh - eg)

a, b, c = matrix[0]
d, e, f = matrix[1]
g, h, i = matrix[2]

determinant = a * (e * i - f * h) - b * (d * i - f * g) + c * (d * h - e * g)

print(f"The 3x3 matrix is:")
for row in matrix:
    print(row)
print(f"\nThe determinant of the matrix is: {determinant}")
```
