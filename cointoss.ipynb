```python
import random

# Number of coin tosses
num_tosses = 1000

# Initialize counts for heads and tails
heads_count = 0
tails_count = 0

# Simulate coin tosses
for _ in range(num_tosses):
    # random.choice(['Heads', 'Tails']) simulates a fair coin toss
    result = random.choice(['Heads', 'Tails'])
    if result == 'Heads':
        heads_count += 1
    else:
        tails_count += 1

# Compute probabilities
probability_heads = heads_count / num_tosses
probability_tails = tails_count / num_tosses

# Display the results
print(f"Total tosses: {num_tosses}")
print(f"Heads count: {heads_count}")
print(f"Tails count: {tails_count}")
print(f"Probability of Heads: {probability_heads:.2f}")
print(f"Probability of Tails: {probability_tails:.2f}")
```
