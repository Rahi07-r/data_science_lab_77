```python
import heapq

def a_star(graph, start, goal, heuristic):
    # Priority queue to store (f_score, node, path)
    # f_score = g_score + h_score
    open_set = []
    heapq.heappush(open_set, (heuristic[start], start, [start]))

    # g_score is the cost from start to current node
    g_score = {node: float('inf') for node in graph}
    g_score[start] = 0

    # came_from stores the path to reconstruct later
    came_from = {}

    while open_set:
        f_score, current_node, path = heapq.heappop(open_set)

        if current_node == goal:
            return path

        for neighbor, weight in graph[current_node].items():
            tentative_g_score = g_score[current_node] + weight

            if tentative_g_score < g_score[neighbor]:
                g_score[neighbor] = tentative_g_score
                came_from[neighbor] = current_node
                f_score = tentative_g_score + heuristic[neighbor]
                new_path = path + [neighbor]
                heapq.heappush(open_set, (f_score, neighbor, new_path))

    return None  # No path found

# Example Usage:
# Define a graph (adjacency list with weights)
graph = {
    'A': {'B': 1, 'C': 4},
    'B': {'A': 1, 'D': 2, 'E': 5},
    'C': {'A': 4, 'F': 2},
    'D': {'B': 2, 'G': 1},
    'E': {'B': 5, 'G': 1},
    'F': {'C': 2, 'G': 3},
    'G': {}
}

# Define heuristic values for each node to the goal 'G'
heuristic = {
    'A': 7,
    'B': 6,
    'C': 2,
    'D': 1,
    'E': 2,
    'F': 1,
    'G': 0
}

start_node = 'A'
goal_node = 'G'

path = a_star(graph, start_node, goal_node, heuristic)

if path:
    print(f"Path from {start_node} to {goal_node}: {path}")
else:
    print(f"No path found from {start_node} to {goal_node}")
```
