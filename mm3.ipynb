```python
# Sample dataset
data = [10, 12, 12, 13, 12, 11, 14, 13, 15, 10, 18, 20, 22, 21, 23]
print(f"Dataset: {data}")

n = len(data)
print(f"Number of elements (n): {n}")

# Calculate Mean
total_sum = 0
for x in data:
    total_sum += x
mean_val = total_sum / n
print(f"Mean: {mean_val:.2f}")
```

```python
# Calculate Median

# Ensure 'data' and 'n' are available from previous cells
# data = [10, 12, 12, 13, 12, 11, 14, 13, 15, 10, 18, 20, 22, 21, 23]
# n = len(data)

sorted_data = sorted(data) # Using built-in sorted function to order the data

if n % 2 == 1:
    # Odd number of elements, median is the middle element
    median_val = sorted_data[n // 2]
else:
    # Even number of elements, median is the average of the two middle elements
    mid1 = sorted_data[n // 2 - 1]
    mid2 = sorted_data[n // 2]
    median_val = (mid1 + mid2) / 2

print(f"Median: {median_val:.2f}")
```

```python
# Calculate Variance

# Ensure 'data', 'n', and 'mean_val' are available from previous cells
# data = [10, 12, 12, 13, 12, 11, 14, 13, 15, 10, 18, 20, 22, 21, 23]
# n = len(data)
# mean_val = total_sum / n

sum_sq_diff = 0
for x in data:
    sum_sq_diff += (x - mean_val) ** 2

# Using sample variance (divide by n-1) for an unbiased estimate
# If n is 1, variance is undefined or 0. Handling this case.
if n > 1:
    variance_val = sum_sq_diff / (n - 1)
else:
    variance_val = 0 # Or handle as an error/undefined

print(f"Variance: {variance_val:.2f}")
```

```python
# Calculate Standard Deviation

# Ensure 'variance_val' is available from previous cells
# variance_val = sum_sq_diff / (n - 1)

std_dev_val = variance_val ** 0.5 # Using power operator for square root

print(f"Standard Deviation: {std_dev_val:.2f}")
```
